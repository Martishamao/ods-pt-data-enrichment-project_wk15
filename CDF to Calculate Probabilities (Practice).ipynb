## Calculate the probability
- Use the normal cumulative distribution function with the mean and standard deviation of female height

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
from scipy import stats
```

```python
filename = 'weight-height.csv'
df = pd.read_csv(filename)
df.head()
```

```python
females = df.loc[df['Gender']=='Female'].copy()
data = females['Height']
```

```python
mean = np.mean(data)
std = np.std(data)
min = np.min(data)
max = np.max(data)
x_range = np.linspace(min, max,1000)
```

```python
def CDF_graph(low_end,high_end):
    cdf = stats.norm.cdf(x_range, loc=mean, scale=std)

    fig, ax = plt.subplots(figsize=(12,6))
    sns.histplot(x=data, stat='density', cumulative = True)
    ax.plot(x_range,cdf)
    ax.axvline(low_end, color= 'black')
    ax.axvline(high_end, color ='black')
    ax.axvspan(low_end, high_end, color ='yellow', zorder = 0)
    ax.set_title('Range of Interest')
    
    probability = stats.norm.cdf(high_end, loc=mean,scale=std) - stats.norm.cdf(low_end,loc=mean, scale=std)
    print(f'Probability: {probability.round(4)}'); 
```

## Between 55.0 and 56.0 inches
- The probability of height between 55 and 56 inches is 0.15%.

```python
CDF_graph(55.0,56.0)
```

## Less than 60.0 inches
- The probability is 8.42%.

```python
CDF_graph(min,60)
```

## Greater than the mean
- The probability that the female height is greater than the mean is 50%.

```python
CDF_graph(mean,max)
```
